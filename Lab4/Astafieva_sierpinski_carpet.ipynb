Намалювати килим Серпинського. 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sierpinski_carpet(n):
    
    # Почнемо з базової матриці, повної одиниць (чорних)
    carpet = np.ones((3**n, 3**n))
    
    # Рекурсивна функція для видалення квадратів
    def remove_squares(size, x, y):
        
        if size == 1:
            return
        
        # Розмір менших квадратів
        new_size = size // 3
        
        # Координати середнього квадрата
        start_x, start_y = x + new_size, y + new_size
        carpet[start_x:start_x+new_size, start_y:start_y+new_size] = 0
        
        # Рекурсія для кожного з 8 навколишніх квадратів
        for dx in range(3):
            for dy in range(3):
                if dx != 1 or dy != 1:  
                    remove_squares(new_size, x + dx * new_size, y + dy * new_size)
    
    
    remove_squares(3**n, 0, 0)
    
    return carpet
```

```python
# Створимо килим Серпінського
order = 8 
carpet = sierpinski_carpet(order)
plt.figure(figsize=(6, 6))
plt.imshow(carpet, cmap='gray')
plt.axis('off')
plt.show()
```
