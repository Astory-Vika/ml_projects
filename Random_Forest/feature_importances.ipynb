```python
#Крок 1. Завантаження набору даних
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
```

```python
#Крок 2. Розподіл даних на тренувальний та тестовий набори
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.2)
```

```python
#Крок 3. Застосування Random Forest для визначення важливих ознак
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, n_jobs=-1)
rf.fit(X_train, y_train)
```

```python
#Крок 4. Побудова графіка, щоб показати важливість кожної ознаки в аналізі
#Використовувається метод feature_importances_ для отримання важливості кожної ознаки

feature_importances = rf.feature_importances_

import matplotlib.pyplot as plt

# Створення графіку важливості ознак
plt.figure(figsize=(10, 6))
plt.bar(range(len(feature_importances)), feature_importances, tick_label=data.feature_names)
plt.title("Важливість ознак в Random Forest")
plt.xticks(rotation=90)
plt.show()
```

```python
#Крок 5. Обговорення результатів та визначення, які ознаки найбільше впливають на вихідні дані
#Як видно з графіка, найбільш важливими ознаками є розмір пухлини, тип пухлини та ядерність ядра
```

```python
#Крок 6. Обчислення точності для двох алгоритмів і порівняння результатів
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Застосування алгоритму SVM
svm = SVC(kernel='linear', C=1)
svm.fit(X_train, y_train)

# Обчислення точності
print("Точність для SVM:", accuracy_score(y_test, svm.predict(X_test)))

# Застосування алгоритму Random Forest
print("Точність для Random Forest:", accuracy_score(y_test, rf.predict(X_test)))
```
